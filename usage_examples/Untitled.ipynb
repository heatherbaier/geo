```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Get all countries with geo data

```python
import requests
from bs4 import BeautifulSoup

# Target directory
url = "https://github.com/Global-Education-Observatory/geo-dataset/tree/master/files_for_db/geo"

# Request and parse
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

# Find all CSV file links
csv_files = [a.text for a in soup.find_all("a") if a.text.endswith(".csv")]

# Extract ISO codes
iso_codes = [f.split("_")[0] for f in csv_files]
print("Found ISO codes:", sorted(set(iso_codes)))
```

## Read in and combine all geospatial data from GitHub

```python
import pandas as pd


geo_data = []
base = "https://raw.githubusercontent.com/Global-Education-Observatory/geo-dataset/master/files_for_db/geo/hnd_geo.csv"

for c, iso in enumerate(iso_codes):
    cur = pd.read_csv(f"https://raw.githubusercontent.com/Global-Education-Observatory/geo-dataset/master/files_for_db/geo/{iso}_geo.csv")
    geo_data.append(cur)
    print(c, len(iso_codes))
    
geo_data = pd.concat(geo_data)
geo_data.head()

# geo = pd.read_csv("https://raw.githubusercontent.com/Global-Education-Observatory/geo-dataset/master/files_for_db/geo/hnd_geo.csv")
# personnel = pd.read_csv("https://raw.githubusercontent.com/Global-Education-Observatory/geo-dataset/master/files_for_db/personnel/hnd_personnel.csv")
# resources = pd.read_csv("https://raw.githubusercontent.com/Global-Education-Observatory/geo-dataset/master/files_for_db/resources/hnd_resources.csv")
```

```python
# Number of schools per country
plt.figure(figsize=(10, 6))
geo_data['adm0'].value_counts().plot(kind='bar')
plt.title('Number of Records per Country')
plt.ylabel('Count')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
