```python
import pandas as pd
import numpy as np
import re

def classify_kgz_school(row):
    name = str(row['school_nam']).strip().lower() if pd.notna(row['school_nam']) else ""
    
    if name == "" or name == "none":
        return "unknown"

    # Pre-primary
    if "kindergarten" in name or "сад" in name or "бакч" in name:
        return "pre_primary"

    # Explicit primary
    if "primary school" in name or "elementary" in name or "pei" in name:
        return "primary"

    # Gymnasiums and Lyceums (upper secondary oriented)
    if "gymnasium" in name or "гимназ" in name or "lyceum" in name or "лицей" in name:
        return "upper_secondary"

    # High schools/general schools
    if "high school" in name or "general high school" in name or re.search(r"\bhs\b", name):
        return "mixed"  # most Kyrgyz HS cover grades 1–11

    # Colleges/universities
    if "college" in name or "университет" in name or "university" in name:
        return "excluded"

    return "unknown"

kgz_ef["school_level"] = kgz_ef.apply(classify_kgz_school, axis = 1)
kgz_ef
```
